```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración para gráficos más grandes
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
# Cargar el archivo ingresos.txt
df = pd.read_csv("ingresos.txt", sep="\t")

# Mostrar las primeras filas
df.head()
```

```python
# Información general del dataset
print("Información del dataset:")
df.info()
```

```python
# Verificar valores nulos
print("\nValores nulos por columna:")
print(df.isnull().sum())
```

```python
# Estadísticas descriptivas
print("\nResumen estadístico:")
display(df.describe())
```

```python
# Imputación de valores nulos

# Imputar peso con la media
df['peso'].fillna(df['peso'].mean(), inplace=True)
```

```python
# Imputar educacion, educacion_madre y educacion_padre con la mediana
df['educacion'].fillna(df['educacion'].median(), inplace=True)
df['educacion_madre'].fillna(df['educacion_madre'].median(), inplace=True)
df['educacion_padre'].fillna(df['educacion_padre'].median(), inplace=True)
```

```python
# Imputar fumador, tenso y malhumorado con la moda
df['fumador'].fillna(df['fumador'].mode()[0], inplace=True)
df['tenso'].fillna(df['tenso'].mode()[0], inplace=True)
df['malhumorado'].fillna(df['malhumorado'].mode()[0], inplace=True)
```

```python
# Verificar que no queden valores nulos
print("Valores nulos después de la imputación:")
print(df.isnull().sum())
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración para gráficos más claros
sns.set(style="whitegrid")
```

```python
# Histograma de ingresos
plt.figure(figsize=(8, 6))
sns.histplot(df['ingresos'], kde=True, bins=30)
plt.title('Distribución de Ingresos')
plt.show()


```

```python
# Aplicar la transformación logarítmica
df['ingresos_log'] = np.log1p(df['ingresos'])

# Visualizar la distribución transformada
plt.figure(figsize=(8, 6))
sns.histplot(df['ingresos_log'], kde=True, bins=30)
plt.title('Distribución Logarítmica de Ingresos')
plt.xlabel('Ingresos (Log)')
plt.show()
```

```python
# Calcular Q1, Q3 e IQR
Q1 = df['ingresos'].quantile(0.25)
Q3 = df['ingresos'].quantile(0.75)
IQR = Q3 - Q1

# Definir límites
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filtrar los valores dentro del rango
df_no_outliers = df[(df['ingresos'] >= lower_bound) & (df['ingresos'] <= upper_bound)]

print("Filas antes de remover outliers:", len(df))
print("Filas después de remover outliers:", len(df_no_outliers))
```

```python

Q1 = df['ingresos'].quantile(0.25)
Q3 = df['ingresos'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR


df_no_outliers = df[(df['ingresos'] >= lower_bound) & (df['ingresos'] <= upper_bound)]


df_no_outliers['ingresos_log'] = np.log1p(df_no_outliers['ingresos'])

plt.figure(figsize=(8, 6))
sns.histplot(df_no_outliers['ingresos_log'], kde=True, bins=30, color='dodgerblue')
plt.title('Distribución de Ingresos (Sin Outliers y con Transformación Log)')
plt.xlabel('Log(Ingresos)')
plt.ylabel('Frecuencia')
plt.show()

```

```python
print(f"Cantidad de valores 0 en ingresos: {len(df[df['ingresos'] == 0])}")
```

```python
# Eliminar filas donde ingresos es igual a 0
df = df[df['ingresos'] > 0]
print(f"Número de filas después de eliminar valores 0: {df.shape[0]}")
```

```python


# Aplicar transformación logarítmica a los ingresos
df['ingresos_log'] = np.log(df['ingresos'])

# Visualización de la distribución final
plt.figure(figsize=(12, 6))

# Distribución original de ingresos
plt.subplot(1, 2, 1)
sns.histplot(df['ingresos'], bins=30, kde=True, color='skyblue')
plt.title('Distribución Final de Ingresos (Original)')
plt.xlabel('Ingresos')
plt.ylabel('Frecuencia')

# Distribución logarítmica de ingresos
plt.subplot(1, 2, 2)
sns.histplot(df['ingresos_log'], bins=30, kde=True, color='lightgreen')
plt.title('Distribución Final de Ingresos (Log Transform)')
plt.xlabel('Log(Ingresos)')
plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```
